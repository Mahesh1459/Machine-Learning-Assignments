```python
from sklearn.preprocessing import MinMaxScaler
Scaler = MinMaxScaler()
d = Scaler.fit_transform(df2)
d
```

```python
df3 = pd.DataFrame(d)
```

```python
df3.columns = [['Murder',"Assault","urban","rape"]]
```

```python
import warnings
warnings.filterwarnings('ignore')
from sklearn.cluster import KMeans
cv = []
for i in range(2, 20):
    km = KMeans(n_clusters=i)
    km.fit_predict(df2)
    cv.append(km.inertia_)
print(cv)
len(cv)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(2,20),cv)
plt.scatter(range(2,20),cv)
plt.show()
```

```python
from sklearn.cluster import KMeans
k = KMeans(n_clusters = 6)
k1 = k.fit_predict(df3)
y_new = pd.DataFrame(k1)
y_new
```

```python
df["KMeans"]= y_new
df
```

```python
#DBSCAN
from sklearn.cluster import DBSCAN
d = DBSCAN(eps = 0.367, min_samples=6)
d.fit(df3)
y = d.labels_
y2 = pd.DataFrame(y)
y2.value_counts()
```
