```python
import pandas as pd
import numpy as np
#load the dataset
df = pd.read_excel('EastWestAirlines.xlsx',sheet_name = "data")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
from sklearn.preprocessing import StandardScaler
Scaler = StandardScaler()
df2 = Scaler.fit_transform(df)
df2
```

```python
df2 = pd.DataFrame(df2)
```

```python
df2
```

```python
import scipy.cluster.hierarchy as shs
dendro_single = shs.dendrogram(shs.linkage(df2,method = "average"))
```

```python
from sklearn.cluster import AgglomerativeClustering
k = AgglomerativeClustering(n_clusters = 3,affinity = 'euclidean' , linkage = 'single' )
k1 = k.fit_predict(df2)
k1
```

```python
df["HCluster"] = k1
```

```python
df["HCluster"].value_counts()
```

```python
from sklearn.cluster import KMeans
c = []
for i in range(2,30):
  v = KMeans(n_clusters = i)
  v.fit_predict(df2)
  c.append(v.inertia_)
print(c)
```

```python
import matplotlib.pyplot as plt
plt.scatter(range(2,30),c)
plt.plot(range(2,30),c)
plt.show()
```

```python
from sklearn.cluster import KMeans
r = KMeans(n_clusters = 14)
r1 = r.fit_predict(df2)
r1
```

```python
df["Kcluster"] = r1
```

```python
df['Kcluster'].value_counts()
```

```python
#DBSCAN
from sklearn.cluster import DBSCAN
o  = DBSCAN(eps = 3, min_samples = 14)
p = o.fit_predict(df2)
p
```

```python
df["DBSCAN Clustering"] = p
```

```python
df["DBSCAN Clustering"].value_counts()
```
