```python
df_ob = df.select_dtypes("object")
df_int = df.select_dtypes("int64")
```

```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
```

```python
for i in df_ob:
  df_ob[i] = LE.fit_transform(df_ob[i])
```

```python
df_ob
```

```python
df5 = pd.concat([df_ob, df_int],axis = 1)
```

```python
df5
```

```python
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
df5 = SS.fit_transform(df5)
```

```python
df5 = pd.DataFrame(df5, columns = [['Undergrad', 'Marital.Status', 'Taxable.Income', 'City.Population',
       'Work.Experience', 'Urban', 'tx']])
```

```python
df5
```

```python
df5 = df5.drop(df5.columns[2], axis = 1)
```

```python
df5
```

```python
X = df5.iloc[:,0:4]
Y = df_ob.iloc[:,-1]
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.25, random_state = 42)
```

```python
print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
```

```python
from sklearn import tree
DTC = tree.DecisionTreeClassifier(criterion = 'gini')
DTC.fit(X_train,Y_train)
```

```python
Y_pred_train = DTC.predict(X_train)
Y_pred_test = DTC.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print('Training Score',accuracy_score(Y_pred_train,Y_train))
print('Testing Score',accuracy_score(Y_pred_test,Y_test))
```

```python
from sklearn.model_selection import GridSearchCV
param = {
    'criterion': ["gini","Entropy"],
    'min_samples_split' : [2,4,6,10],
    'min_samples_leaf' :[2,6,8,12],
    'max_features': ["auto","sqrt","log2"],
    'max_depth' : [4,6,8,12,10]

}
grid = GridSearchCV(DTC,param)
grid.fit(X_train,Y_train)
grid.score(X_train,Y_train)
print(grid.best_score_)
print(grid.best_params_)
```
