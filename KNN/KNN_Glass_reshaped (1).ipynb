```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("glass.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
df1 = SS.fit_transform(df)
df1 = pd.DataFrame(df1, columns = [['RI', 'Na', 'Mg', 'Al', 'Si', 'K', 'Ca', 'Ba', 'Fe', 'Type']])
```

```python
df1
```

```python
X = df1.iloc[:,:-1]
Y  = pd.DataFrame(df["Type"])
```

```python
Y.value_counts()
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.25, random_state = 42)
```

```python
print(X_train.shape, Y_train.shape, X_test.shape, Y_test.shape)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
```

```python
param_Grid = dict(n_neighbors = np.array(range(2,20)))
```

```python
Y
```

```python
model = KNeighborsClassifier()
grid = GridSearchCV(estimator = model, param_grid = param_Grid)
```

```python
grid.fit(X,Y)
```

```python
print(grid.best_score_)
print(grid.best_params_)
```

```python
KNN = KNeighborsClassifier(n_neighbors = 3)
KNN.fit(X_train,Y_train)
KNN.score(X_train,Y_train)
```

```python
Y_pred_train1 = KNN.predict(X_train)
Y_pred_test1 = KNN.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print("Trainning Accuracy", accuracy_score(Y_pred_train1, Y_train).round(2))
print("Testing Accuracy", accuracy_score(Y_pred_test1,Y_test).round(2))
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_pred_train1,Y_train))
```

```python
from imblearn.over_sampling import SMOTE
n_neighbors = 3
# Apply SMOTE with the specified n_neighbors
smote = SMOTE(sampling_strategy='auto', k_neighbors=n_neighbors, random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, Y)
```

```python
from sklearn.model_selection import train_test_split
X_train_reshaped,X_test_reshaped,Y_train_reshaped,Y_test_reshaped = train_test_split(X_resampled,y_resampled,test_size = 0.3,random_state = 42)
```

```python
print(Y_train_reshaped.shape)
```

```python
n_neighbors = np.array(range(2,10))
param_grid = dict(n_neighbors = n_neighbors)
```

```python
model = KNeighborsClassifier()
grid = GridSearchCV(estimator = model, param_grid = param_grid)
grid.fit(X_train_reshaped,Y_train_reshaped)
```

```python
print(grid.best_score_)
print(grid.best_params_)
```

```python
from sklearn.neighbors import KNeighborsClassifier
KNN = KNeighborsClassifier(n_neighbors = 2)
KNN.fit(X_train_reshaped,Y_train_reshaped)
KNN.score(X_train_reshaped,Y_train_reshaped).round(2)
```

```python
Y_pred_train1 = KNN.predict(X_train_reshaped)
Y_pred_test1 = KNN.predict(X_test_reshaped)
```

```python
from sklearn.metrics import accuracy_score
print("Trainning Accuracy", accuracy_score(Y_pred_train1, Y_train_reshaped).round(2))
print("Testing Accuracy", accuracy_score(Y_pred_test1,Y_test_reshaped).round(2))
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_pred_train1,Y_train_reshaped))
```

```python
from sklearn.model_selection import KFold,cross_val_score
k = KFold( n_splits= 5)
result = cross_val_score(KNN,X_train_reshaped,Y_train_reshaped,cv = k)
```

```python
print(np.mean(result).round(2))
```
