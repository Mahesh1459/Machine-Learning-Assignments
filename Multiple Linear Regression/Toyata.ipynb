# Transformations

```python
model_log = smf.ols("Price~np.log(Age)+np.log(KM)+np.log(HP)+np.log(CC)+np.log(Doors)+np.log(Gears)+np.log(Q)+np.log(Weight)",data = df1).fit()
model_log.summary()
```

```python
model_log.rsquared
```

```python
model_sqrt = smf.ols("Price~np.sqrt(Age)+np.sqrt(KM)+np.sqrt(HP)+np.sqrt(CC)+np.sqrt(Doors)+np.sqrt(Gears)+np.sqrt(Q)+np.sqrt(Weight)",data = df1).fit()
model_sqrt.summary()
```

```python
model_sqrt.rsquared
```

```python
df = {"Model" : pd.Series(["model","model2","model_log","model_sqrt"]),
     "R squared values" : pd.Series([model.rsquared,model2.rsquared,model_log.rsquared,model_sqrt.rsquared])
     }
data = pd.DataFrame(df)
data
```

```python
df1.head()
```

```python
#model is before the multi collinearity issues
#model2 is after the removing of influence points and resolving the multi collinearity issues
#model_log is used for Transformation
#model_Sqrt is square root transformation
#Best model is the Second model
```
