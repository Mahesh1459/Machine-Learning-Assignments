```python
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("SalaryData_Train.csv")
df
```

```python
df_object = df.select_dtypes('object')
df_numerical = df.select_dtypes('int64')
```

```python
df_object
```

```python
from sklearn.preprocessing import LabelEncoder
for i in df_object.iloc[:,:-1]:
    LE = LabelEncoder()
    df_object[i] = LE.fit_transform(df_object[i])
```

```python
df_object
```

```python
df_new_train = pd.concat([df_numerical, df_object],axis = 1)
df_new_train
```

```python
#split your train data into X and Y
X_train = df_new_train.iloc[:,:-1]
Y_train = df_new_train.iloc[:,-1]
```

```python
X_train
```

```python
Y_train
```

```python
#Load the test data
df_test = pd.read_csv('SalaryData_Test.csv')
df_test
```

```python
df_test.head()
```

```python
df_test.columns
```

```python
df_test.dtypes
```

```python
df_test_object = df_test.select_dtypes("object")
df_test_numerical = df_test.select_dtypes("int64")
```

```python
from sklearn.preprocessing import LabelEncoder
for i in df_test_object.iloc[: :-1]:
    LE = LabelEncoder()
    df_test_object[i] = LE.fit_transform(df_test_object[i])
```

```python
df_new_test = pd.concat([df_test_numerical, df_test_object],axis =1)
df_new_test
```

```python
X_test = df_new_test.iloc[:,:-1]
Y_test = df_new_test.iloc[:,-1]
```

```python
print(X_test.shape,Y_test.shape)
```

```python
from sklearn.naive_bayes import MultinomialNB
MNB1 = MultinomialNB()
MNB1.fit(X_train,Y_train)
MNB1.score(X_train,Y_train)
```

```python
from sklearn.naive_bayes import MultinomialNB
MNB2 = MultinomialNB()
MNB2.fit(X_test,Y_test)
MNB2.score(X_test,Y_test)
```

```python
Y_pred_train = MNB1.predict(X_train)
Y_pred_test = MNB2.predict(X_test)
```

```python
#Accurarcy metrics
from sklearn.metrics import accuracy_score
print('Training Accuracy',accuracy_score(Y_pred_train,Y_train).round(2))
print('Testing Accuracy',accuracy_score(Y_pred_test,Y_test).round(2))
```
