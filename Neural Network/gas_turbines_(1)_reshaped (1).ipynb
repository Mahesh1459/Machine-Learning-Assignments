```python

from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
```

```python
import pandas as pd
df = pd.read_csv("gas_turbines.csv")
df.head()
```

```python
X = df.drop(df.columns[7],axis = 1)
X
```

```python


Y = df['TEY']
```

```python
model = Sequential()
model.add(Dense(12, input_dim=10,  activation='relu')) #input layer
model.add(Dense(1, activation='linear')) #output layer
```

```python

model.compile(loss='mean_squared_error', optimizer='adam', metrics=['mean_squared_error'])
```

```python
history = model.fit(X, Y, validation_split=0.33, epochs=50, batch_size=10)
```

```python
model.evaluate(X,Y)
```

```python
model.add(Dense(100,activation = 'relu',input_shape = (10,)))
model.add(Dense(50,activation = 'relu'))
model.add(Dense(1,activation = 'linear'))
```

```python
model.compile(optimizer = 'adam',loss = 'mse',metrics = ['mean_squared_error'])
```

```python
model_ann = model.fit(X,Y,epochs = 100)
model_ann
```

```python
model.evaluate(X,Y)
```

```python
#MSE Graph
import matplotlib.pyplot as plt
plt.plot(model_ann.history['mean_squared_error'],'r')
plt.title('MSE Graph')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
```

```python
import matplotlib.pyplot as plt
plt.plot(model_ann.history['loss'],'b')
plt.title('Loss Graph')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
```
