```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("Company_Data.csv")
df.head()
```

```python
df.info()
```

```python
df_object = df.select_dtypes('object')
df_object
```

```python
df.describe()
```

```python
df_object
```

```python
from sklearn.preprocessing import LabelEncoder
for i in df_object:
  LE = LabelEncoder()
  df_object[i] = LE.fit_transform(df_object[i])
```

```python
df_object
```

```python
df_int = df.select_dtypes("int64")
df_float = df.select_dtypes("float")
```

```python
df = pd.concat([df_object,df_int,df_float],axis = 1)
```

```python
df.columns
```

```python
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
df2 = SS.fit_transform(df)
```

```python
df2 = pd.DataFrame(df2, columns = [['ShelveLoc', 'Urban', 'US', 'CompPrice', 'Income', 'Advertising',
       'Population', 'Price', 'Age', 'Education', 'Sales']])
```

```python
df2
```

```python
X = df2.iloc[:,:-1]
Y = df["Sales"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.3,random_state = 42)
```

```python
print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn import tree
RFR = RandomForestRegressor()
RFR.fit(X_train,Y_train)
```

```python
RFR.score(X_train,Y_train)
```

```python
from sklearn.model_selection import RandomizedSearchCV
n_estimators = [100,200,300,400,500]
max_samples = [0.4,0.5,0.6,0.7,0.8]
max_features = [0.4,0.5,0.6,0.7,0.8]
min_samples_split = [4,5,6,7,8]
min_samples_leaf = [4,5,6,7,8]
params = {
    'n_estimators':n_estimators,
    'max_samples':max_samples,
    'max_features':max_features,
    'min_samples_split':min_samples_split,
    'min_samples_leaf':min_samples_leaf,
}
descision_tree_new = RandomizedSearchCV(estimator = RFR, param_distributions =params,n_jobs =-1,n_iter  = 200,cv =5,random_state = 42)
descision_tree_new.fit(X_train,Y_train)
```

```python
descision_tree_new.best_params_
```

```python
descision_tree_new.best_estimator_
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn import tree
RFR = RandomForestRegressor(n_estimators = 100, min_samples_split = 8,min_samples_leaf =4, max_samples =0.8,max_features =0.5)
RFR.fit(X_train,Y_train)
```

```python

Y_pred_train = RFR.predict(X_train)
Y_pred_test = RFR.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
print('Training Error',np.mean(mean_squared_error(Y_pred_train,Y_train)).round(2))
print('Testing Error',np.mean(mean_squared_error(Y_pred_test,Y_test)).round(2))
```

```python
from sklearn.model_selection import KFold,cross_val_score
k = KFold( n_splits= 5)
result = cross_val_score(RFR,X,Y,cv = k)
```
