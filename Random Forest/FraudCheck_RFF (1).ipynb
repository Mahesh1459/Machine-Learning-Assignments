```python
new = []
for i in df["Taxable.Income"]:
  if i <=30000:
    new.append("Risky")
  else:
    new.append("Good")
```

```python
df["tx"] = new
```

```python
df_ob = df.select_dtypes("object")
df_int = df.select_dtypes("int64")
```

```python
df_int
```

```python
df_ob
```

```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
```

```python
for i in df_ob:
  df_ob[i] = LE.fit_transform(df_ob[i])
```

```python
df_ob
```

```python
df5 = pd.concat([df_ob, df_int],axis = 1)
```

```python
df5
```

```python
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
df5 = SS.fit_transform(df5)
```

```python
df5 = pd.DataFrame(df5, columns = [['Undergrad', 'Marital.Status', 'Taxable.Income', 'City.Population',
       'Work.Experience', 'Urban', 'tx']])
```

```python
df5
```

```python
df5 = df5.drop(df5.columns[2], axis = 1)
```

```python
df5
```

```python
X = df5.iloc[:,0:4]
Y = df_ob.iloc[:,-1]
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.25, random_state = 42)
```

```python
print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
```

```python
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
RFC = RandomForestClassifier(n_estimators = 100, criterion = "entropy", max_samples =0.5, max_features = 0.6, min_samples_split = 5)
RFC.fit(X_train,Y_train)
```

```python
Y_pred_train = RFC.predict(X_train)
Y_pred_test = RFC.predict(X_test)
```

```python
Y_pred_test
```

```python
Y_test
```

```python
from sklearn.metrics import accuracy_score
print('Training Score',accuracy_score(Y_pred_train,Y_train))
print('Testing Score',accuracy_score(Y_pred_test,Y_test))
```

```python
#checking the hyper parameter optimization for checking the best parameter
from sklearn.model_selection import RandomizedSearchCV
n_estimators = [100,200,300,400,500,600]
criterion = ['entropy','gini']
max_samples = [0.5,0.6,0.7,0.8,0.9]
max_features = [0.5,0.6,0.7,0.8,0.9]
min_samples_split = [4,5,6,7,8]
value = {
    'n_estimators':n_estimators,
    'criterion':criterion,
    'max_samples':max_samples,
    'max_features':max_features,
    'min_samples_split':min_samples_split
}
RSCV = RandomizedSearchCV(RFC,cv  = 5,param_distributions = value,n_iter = 10,n_jobs = -1)
RSCV.fit(X_train,Y_train)
print(RSCV)
```

```python
RSCV.best_params_
```

```python
from sklearn.model_selection import KFold,cross_val_score
K = KFold(20)
result = cross_val_score(RFC,X,Y,cv = K)
print(np.mean(result))
```
