```python
df_new_object = df.select_dtypes('object')
df_new_numerical =  df.select_dtypes(['int64','float64'])
```

```python
df_new_numerical.describe()
```

```python
from sklearn.preprocessing import LabelEncoder
for i in df_new_object:
    LE = LabelEncoder()
    df_new_object[i] = LE.fit_transform(df_new_object[i])
```

```python
#combine the numerical and categorical datatype
df_new_data = pd.concat([df_new_numerical,df_new_object],axis =1)
df_new_data
```

```python
from sklearn.preprocessing import StandardScaler
for i in df_new_data:
    SS = StandardScaler()
    df_new_data = SS.fit_transform(df_new_data)
```

```python
column = ['month', 'day', 'FFMC', 'DMC', 'DC', 'ISI', 'temp', 'RH', 'wind',
       'rain', 'area', 'dayfri', 'daymon', 'daysat', 'daysun', 'daythu',
       'daytue', 'daywed', 'monthapr', 'monthaug', 'monthdec', 'monthfeb',
       'monthjan', 'monthjul', 'monthjun', 'monthmar', 'monthmay', 'monthnov',
       'monthoct', 'monthsep','size_category']
df_new_data = pd.DataFrame(df_new_data,columns = column)
```

```python
X = df_new_data.drop('size_category',axis = 1)
Y = df['size_category']
```

```python
#Data Partition 
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.3,random_state = 42)
```

```python
print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
```

```python
#Data modeling
from sklearn.svm import SVC
S = SVC()
S.fit(X_train,Y_train)
S.score(X_train,Y_train)
```

```python
Y_pred_train = S.predict(X_train)
Y_pred_test = S.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
print('Training Accuracy',accuracy_score(Y_pred_train,Y_train).round(2))
print('Testing Accuracy',accuracy_score(Y_pred_test,Y_test).round(2))
```

```python
#For doing the hyper parameter tuning for getting better accuarcy
from sklearn.model_selection import GridSearchCV
kernel = ['linear','ploy','rbf','sigmoid']
C = [1,2,3,4]
gamma = [10,20,30]
param = {
    'kernel':kernel,
    'C':C,
    'gamma':gamma
}
result = GridSearchCV(estimator = S,param_grid = param,n_jobs =-1,cv = 5)
result.fit(X_train,Y_train)
print(result)
```

```python
result.best_params_
```

```python
from sklearn.svm import SVC
S = SVC(kernel  = 'linear',C=4,gamma = 10)
S.fit(X_train,Y_train)
S.score(X_train,Y_train)
```

```python
Y_pred_train = S.predict(X_train)
Y_pred_test = S.predict(X_test)
from sklearn.metrics import accuracy_score,classification_report
print('Training Accuracy',accuracy_score(Y_pred_train,Y_train).round(2))
print('Testing Accuracy',accuracy_score(Y_pred_test,Y_test).round(2))
```

```python
#cross validation
from sklearn.model_selection import KFold,cross_val_score
K = KFold(n_splits = 5)
result = cross_val_score(S,X,Y,cv = K)
print(result)
```

```python
#mean for the KFold cross validation
print(np.mean(result).round(2))
```
